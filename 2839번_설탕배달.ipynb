# 문제
상근이는 요즘 설탕공장에서 설탕을 배달하고 있다. 상근이는 지금 사탕가게에 설탕을 정확하게 N킬로그램을 배달해야 한다. 설탕공장에서 만드는 설탕은 봉지에 담겨져 있다. 봉지는 3킬로그램 봉지와 5킬로그램 봉지가 있다.

상근이는 귀찮기 때문에, 최대한 적은 봉지를 들고 가려고 한다. 예를 들어, 18킬로그램 설탕을 배달해야 할 때, 3킬로그램 봉지 6개를 가져가도 되지만, 5킬로그램 3개와 3킬로그램 1개를 배달하면, 더 적은 개수의 봉지를 배달할 수 있다.

상근이가 설탕을 정확하게 N킬로그램 배달해야 할 때, 봉지 몇 개를 가져가면 되는지 그 수를 구하는 프로그램을 작성하시오.

# 입력
첫째 줄에 N이 주어진다. (3 ≤ N ≤ 5000)

# 출력
상근이가 배달하는 봉지의 최소 개수를 출력한다. 만약, 정확하게 N킬로그램을 만들 수 없다면 -1을 출력한다.

## 풀이1
1. 설탕 무게를 입력받는다.
2. 5킬로그램의 봉지가 최대한 많아야 봉지의 개수가 적다.
3. 설탕 무게를 5로 나눈 나머지를 또 3으로 나눠서 나머지가 0이라면 이는 5킬로그램 봉지를 최대한 활용하여 가져갈 수 있다는 의미와 같다.
**이 때, 11과 같이 5킬로그램을 2봉지를 가져가면 3킬로그램으로 챙길 수 없다고 판단할 수 있다. 하지만 5킬로그램 1봉지 3킬로그램 2봉지로 가져갈 수 있다. 이를 고려하기 위해서는 5킬로그램 봉지 수를 줄여가면서 3킬로그램으로 나머지를 가져갈 수 있는지를 판단하는 조건이 필요하다.**

4. 위 조건은 5킬로그램 봉지로만 가져가는 것도 포함하며, 위 조건이 만족하지 않으면 3킬로그램 봉지로만 가져가는 방법이 있는지 확인한다.
5. 위 두 조건이 모두 만족하지 않는다면 -1을 출력한다.

```python
# 설탕의 무게 입력
weight = int(input())

# 봉지의 개수(최종 결과)
result = 0


# 5킬로그램 봉지, 3킬로그램 봉지 혼합하여 가져갈 수 있거나, 5킬로그램 봉지로만 가져갈 수 있는 경우, 단 이 때는 5킬로그램 봉지를
# 최대한 활용할 때만 성립한다.
if (weight % 5) % 3 == 0: 
    result = int(weight / 5) # 필요한 5킬로그램 봉지 개수 더하기
    result = result + int((weight - (result * 5)) / 3) # 필요한 3킬로그램 봉지 개수 더하기
    print(result)
    
# 3킬로그램 봉지만 활용해야하거나, 5킬로그램 봉지를 줄이면 3킬로그램 봉지로 다 가져갈 수 있는 경우
else:
    temp = int(weight / 5)
    check = 0 # 모두 성립하지 않는 경우를 판단하는 지표
    for i in range(temp):
        if (weight - int((temp - (i + 1)) * 5)) % 3 == 0:
            result = temp - (i + 1)
            result = result + int((weight - (result * 5)) / 3) 
            print(result)
            break
        else:
            check += 1
    if(check == temp): # 루프를 다 돌아도 성립하는 경우를 못 찾은 경우
        print(-1)
```

## 풀이 2
1. 설탕 무게를 입력 받는다.
2. 설탕 무게를 5로 나눈 몫을 5킬로그램 봉지 개수로 초기화 한다.
3. 설탕 무게를 5로 나눈 나머지를 구한다.
4. 5로 나눈 몫이 0보다 클 동안
    4-1 5로 나눈 나머지가 3으로 나눠지면
        5로 나눈 나머지를 3으로 나눈 몫을 구하고 이는 3킬로그램 봉지의 개수가 된다.
    4-2 5로 나눈 나머지가 3으로 나눠지지 않으면
        5로 나누는 몫을 하나 뺀다.
        5로 나눈 나머지에 5를 더한다.
    반복
5. 3으로 나눠졌다면 5로 나눈 몫과 3으로 나눈 몫을 더해 결과를 출력한다.
5. 3으로 나눠지지 않았다면 -1을 출력한다.

```python
weight = int(input()) # 설탕무게 입력
a = weight // 5 # 5킬로그램 봉지 개수 초기화
b = weight % 5 # 5로 나눈 나머지
c = 0 # 3킬로그램 봉지 개수 초기화

while a >= 0: # 5로 나눈 몫이 0 이상인 경우 반복
    if b % 3 == 0: # 5킬로그램 봉지와 3킬로그램 봉지로 가져갈 수 있는 경우
        c = b // 3 # 3킬로그램 봉지의 개수 구하기
        break # 찾으면 이때가 최소로 봉지를 가져갈 수 있으므로 중단해야한다.
    else: # 딱 떨어지지 않는 경우
        a -= 1 # 5킬로그램 봉지 개수를 줄인다.
        b += 5 # 5킬로그램 봉지로 가져간 나머지에서 5를 더한다.
        
if b % 3 == 0: # 3, 5 킬로그램 봉지로 가져갈 수 있는 경우
    print(a + c) # 봉지 개수 더해서 결과 출력
else: # 가져갈 수 없는 경우
    print(-1) # -1 출력
```
